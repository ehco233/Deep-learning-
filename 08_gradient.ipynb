


## 梯度爆炸
- 值超出 值域 infinity
  - 16 位浮点数尤为严重
- 对学习率非常敏感
  - 学习率太大，大参数值，更大的梯度
  - 学习率太小，训练无进展
  - 需要不断的调整学习率

## 梯度消失
- 梯度变成0
  - 16位浮点数尤其严重
- 训练没有进展
- 底层尤为严重
  - 仅仅顶部层训练的比较好，无法让nn更深
  
## 让训练更加稳定
- 目标：将梯度在合理的范围内
  - 例如 [1e-6 , 1e3]
- 将乘法变加法
  - ResNet
  - LSTM
- 归一化
  - 梯度归一化，梯度裁剪
- 合理的权重初始化和激活函数
